# Exploratory Data Analysis in Pandas

```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load datasets
df = pd.read_csv("/content/world_population.csv")
df
```

```python
pd.set_option('display.float_format', lambda x : '%.2f'% x)
df
```

```python
#looking at data
df.info()
```

```python
df.describe()
```

```python
#look for null values
df.isnull().sum()
```

```python
#unique values
df.nunique()
```

```python
#sorting the data
df.sort_values(by = '2022 Population', ascending = False).head(10)
```

```python
df.sort_values(by = 'World Population Percentage', ascending = False).head(10)
```

```python
#correlation between columns
df.corr()
```

```python
#heatmap of corr matrix
sns.heatmap(df.corr(), annot = True)
plt.rcParams['figure.figsize'] = (10,5)
plt.show()
```

```python
#grouping the data
df.groupby('Continent').mean()
```

```python
#what is oceania?
df[df['Continent'].str.contains('Oceania')]
```

```python
df.groupby('Continent').mean().sort_values(by='2022 Population', ascending = False)
```

```python
df.columns
```

```python
#group only population columns
df2 = df.groupby('Continent')['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population'].mean().sort_values(by='2022 Population', ascending = False)
```

```python
df2
```

```python
df2.plot()
```

The graph doesnot feel right as it is plotting all population values on continent as X-axis.

```python
#transpose the grouped data
df3 = df2.transpose()
df3
```

```python
df3.plot(figsize =(20,5))
```

Its looks good but line is plotted backwards.

```python
df4 = df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by='2022 Population', ascending = False)
```

```python
df5 = df4.transpose()
df5
```

```python
df5.plot(figsize =(20,5))
```

```python
#boxplots for outliers
df.boxplot(figsize = (20,10))
```

In our datasets, it is expected to have outliers or high differences of values, but in case of other datasets we need to inspect those outlier values.
