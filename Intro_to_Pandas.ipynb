**DataFrames:**

Let's create dataframes of car price data that we can use in this lesson.

```python
#creating dataframes from list
data = [
        ['Nissan', 'Stanza', 1991, 138, 4, 'MANUAL', 'sedan', 2000],
        ['Hyundai', 'Sonata', 2017, None, 4, 'AUTOMATIC', 'Sedan', 27150],
        ['Lotus', 'Elise', 2010, 218, 4, 'MANUAL', 'convertible', 54990],
        ['GMC', 'Acadia',  2017, 194, 4, 'AUTOMATIC', '4dr SUV', 34450],
        ['Nissan', 'Frontier', 2017, 261, 6, 'MANUAL', 'Pickup', 32340]
    ]

columns = ['Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders',
           'Transmission Type', 'Vehicle_Style', 'MSRP']
```

```python
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data,columns = columns)
df
```

**Indexing and Accessing the elements**

```python
df.index
```

```python
df.Make.index
```

```python
#to access the data from index
df.loc[1] #passing single index
```

```python
df.loc[[1,2]] #passing multiple index
```

```python
#changing the index
df.index = ['a','b','c','d','e']
df
```

```python
df.loc[[1,2]] #no longer acccess the elements
```

```python
df.loc[['a','b']]
```

```python
#to access the elements with index when we have changed the index
df.iloc[[1,2]]
```

```python
#reset the index
df =df.reset_index(drop = True)
df
```

**Element wise Operations:**

```python
df["Engine HP"] /100
```

```python
df['Year'] >= 2015
```

**Filtering:**

```python
df[
    df['Year'] >= 2015
    ]
```

```python
df[
    df['Make'] == 'Nissan'
]
```

```python
df[
    (df['Year'] >= 2015) & (df['Make']=='Nissan')
]
```
