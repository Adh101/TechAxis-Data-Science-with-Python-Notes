# Filtering and Ordering

```python
data = pd.read_csv('/content/world_population.csv')
data
```

```python
#filtering with specific column
data[data['Rank'] <= 10]
```

```python
specific_countries = ['Brazil','Bangladesh']
```

```python
data[data['Country'].isin(specific_countries)]
```

```python
data[data['Country'].str.contains('Nepal')]
```

```python
#setting index
data = data.set_index('Country')
```

```python
data
```

```python
data.filter(items=['Rank','CCA3','Capital'],axis=1) #setting axis to 1 search in column axis
```

```python
data.filter(items=['Nepal'], axis=0) #setting axis to 0 search in row axis
```

```python
data.filter(like = 'United', axis = 0)
```

```python
data[data['Rank'] < 10]
```

```python
#sorting the values
data[data['Rank'] < 10].sort_values(by='Rank')
```

```python
data[data['Rank'] < 10].sort_values(by='Rank', ascending = False)
```

```python
#sorting by multiple columns
data[data['Rank'] < 10].sort_values(by=['Continent','Country'],ascending = True) #first continent then country
```

```python
data[data['Rank'] < 10].sort_values(by=['Country','Continent'],ascending = True) #first country then continent
```

```python
data[data['Rank'] < 10].sort_values(by=['Continent','Country'],ascending = [False,True])
```

# Group by and Aggregate Functions

```python
group_by_con = df.groupby('Continent')
```

```python
group_by_con.mean()
```

```python
df.groupby('Continent').mean()
```

```python
df.groupby('Continent').count()
```

```python
df.groupby('Continent').sum()
```

```python
df.groupby('Continent').describe()
```

```python
#aggregate fucntions
df.groupby('Continent').agg({'2022 Population':['mean','max','count','sum']})
```

```python
df.groupby(['Continent','Country']).agg({'2022 Population':['mean','max','count','sum']})
```
