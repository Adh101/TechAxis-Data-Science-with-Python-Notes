##t-Distributed Stochastic Neighbor Embedding (t-SNE):

t-Distributed Stochastic Neighbor Embedding (t-SNE) is a non-linear dimensionality reduction technique particularly useful for visualizing high-dimensional data in a lower-dimensional space. It is effective in capturing both global and local structures of the data.

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
model = TSNE(n_components= 2, perplexity=50, n_iter = 1000,random_state=42)
X_tsne = model.fit_transform(X)
```

```python
X_tsne.shape
```

```python
X.shape
```

```python
plt.figure(figsize=(8, 6))
for i in range(len(np.unique(y))):
        plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=f'Class {i}')

plt.title('t-SNE of Iris Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.show()
```

In this example, t-SNE is applied to the Iris dataset to demonstrate its ability to capture the underlying structure of high-dimensional data in a 2D space.
