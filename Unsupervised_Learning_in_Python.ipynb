##Unsupervised Learning in Python:

- It is used to discover patterns in unlabeled data.
- Unsuperised learning finds pattern in data without considering specific prediction task like supervised learning.
- Eg: *clustering* customers by their purchases, compressing the data using purchasing patterns (*dimensionality reduction*)

####k-Means Clustering:
- finds clusters of samples,
- no of clusters must be specified,


          from sklearn.cluster import KMeans
          
          model = KMeans(n_clusters = n)

          model.fit(data)

          model.predict(data)

Algorithm Steps:

1. Initialization:
  -Select values of k initial cluster centroids,
2. Assignment:
 - Assign each data point to the nearest centroid forming k-clusters,

3. Update Centroids:
  - Recalculate the cemtroid of each cluster as the mean of data points assigned to it.
4. Repeat steps 2 and 3 until convergence( when centroids stop changing) or maximum iterations is reached.

```python
from sklearn.datasets import load_iris
```

```python
#loading iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
#importing the KMeans
from sklearn.cluster import KMeans

#creating instance of model
model = KMeans(n_clusters = 3)

#fitting model on data
model.fit(X)
```

```python
#predicting the labels
labels = model.predict(X)
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Standardize the feature data (necessary for K-Means)
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Choose the number of clusters (K)
k = 3  # As there are three species of Iris in the dataset

# Initialize the K-Means model
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the K-Means model to the standardized data
kmeans.fit(X_std)

# Get cluster labels for each data point
labels = kmeans.labels_

# Add the cluster labels to the original dataset
iris['Cluster'] = labels

# Visualize the clusters
plt.figure(figsize=(10, 5))

# Scatter plot for sepal length and sepal width
plt.subplot(1, 2, 1)
plt.scatter(X_std[:, 0], X_std[:, 1], c=labels, cmap='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (Standardized)')
plt.ylabel('Sepal Width (Standardized')

# Scatter plot for petal length and petal width
plt.subplot(1, 2, 2)
plt.scatter(X_std[:, 2], X_std[:, 3], c=labels, cmap='viridis')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (Standardized)')
plt.ylabel('Petal Width (Standardized')

plt.show()
```
