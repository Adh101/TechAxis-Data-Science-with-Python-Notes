## Web Scraping in Python:

```python
from bs4 import BeautifulSoup

import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
#find the table
soup.find('table')
```

```python
soup.find_all('table')  #returns all table from the page
```

```python
soup.find_all('table', class_ = "wikitable sortable")
```

```python
table = soup.find_all('table')[1]
```

```python
titles = table.find_all('th')
```

```python
table_titles = [title.text.strip() for title in titles]
print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=table_titles)
df
```

```python
#find data
row_data = table.find_all('tr')
```

```python
for row in row_data:
  print(row.find_all('td'))
```

```python
for row in row_data:
  column_data = row.find_all('td')
  individual_data = [data.text.strip()for data in column_data]
  print(individual_data)
```

```python
len(df)
```

```python
for row in row_data[1:]:
  column_data = row.find_all('td')
  individual_data = [data.text.strip()for data in column_data]

  length = len(df)
  df.loc[length] = individual_data
```

```python
table = soup.find_all('table')[1]
titles = table.find_all('th')

table_titles = [title.text.strip() for title in titles]

df = pd.DataFrame(columns=table_titles)
```

```python
#find data
row_data = table.find_all('tr')

for row in row_data[1:]:
  column_data = row.find_all('td')
  individual_data = [data.text.strip()for data in column_data]

  length = len(df)
  df.loc[length] = individual_data

df
```
