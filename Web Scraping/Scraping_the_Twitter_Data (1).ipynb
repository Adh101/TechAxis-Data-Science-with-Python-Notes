```python
import tweepy
```

```python
# Replace with your API keys and tokens (use your own key from developer.twitter.com)
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
# Authenticate
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
hashtag = 'imVkohli'
tweet_count = 10  to

tweets = tweepy.Cursor(api.search_tweets, q=hashtag, count=tweet_count, lang='en', tweet_mode='extended').items()


for tweet in tweets:
    print(f"Tweet: {tweet.full_text}")
    print('---')
```

```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
tweets = scraper.get_tweets('palestine',mode ='hashtag', number = 10)
```

```python
tweets
```

```python
final_tweets = []

for tweet in tweets['tweets']:
  data = [tweet['link'],tweet['text'],tweet['date'],tweet['stats']['likes'],tweet['stats']['comments']]
  final_tweets.append(data)
```

```python
final_tweets
```

```python
df = pd.DataFrame(final_tweets,columns = ['link','text','date','likes','comments'])
df
```

```python
def tweet_scraper(name,modes, no):
    tweets_new = scraper.get_tweets(name,mode =modes, number = no)
    final_tweet = []

    for tweet in tweets_new['tweets']:
      data = [tweet['link'],tweet['text'],tweet['date'],tweet['stats']['likes'],tweet['stats']['comments']]
      final_tweet.append(data)

    df = pd.DataFrame(final_tweet,columns = ['link','text','date','likes','comments'])
    return df
```

```python
tweets_scraped = tweet_scraper('palestine','hashtag',10)
```

```python
tweets_scraped
```

```python
tweets_scraped.to_csv('tweets.csv')
```
