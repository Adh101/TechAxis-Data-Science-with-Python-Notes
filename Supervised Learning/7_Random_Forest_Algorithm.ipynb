##Random Forest Algorithm:
- It is the supervised learning algorithms for both classification and regression tasks.
- It uses Ensemble Method i.e. combines predictions from multiple models.
- In Random Forest, it is the combination(ensemble) of multiple decision tress.
- It builds the multiple decision trees and merges them to get more accurate and stable predictions.

####Data Preprocessing:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
plt.gray()
for i in range(10):
  plt.matshow(digits.images[i])
```

```python
#create dataframe
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

####Model Training and Prediction

```python
X = df.drop('target',axis ='columns')
y = df.target
```

```python
#split the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3,
                                                    random_state = 42)
```

```python
#model

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 20)

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.score(X_test,y_test)
```

#### Model Evaluation:

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)

cm
```

```python
#trees
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz

for i in range(3):
  tree = model.estimators_[i]
  dot_data = export_graphviz(tree, feature_names = X_train.columns,
                             filled = True,
                             max_depth = 2,
                             impurity = False,
                             proportion = True)

  graph = graphviz.Source(dot_data)
  display(graph)

```

```python
#visaulizing confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot = True)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
#hyperparameter tuning

from scipy.stats import randint

from sklearn.model_selection import RandomizedSearchCV

#parameters
param_dist = {'n_estimators': randint(1,100),
              'max_depth': randint(1,20)}

#model

rf = RandomForestClassifier()

#using randomized search

rand_search = RandomizedSearchCV(rf,
                                 param_distributions = param_dist,
                                 n_iter = 5,
                                 cv = 5)

#fit model
rand_search.fit(X_train,y_train)

```

```python
#finding best hyperparameters
best_parameters = rand_search.best_params_

print('Best Hyperparamters:',best_parameters)
```

```python
rand_search.best_estimator_
```

```python
#applying best parameters to model
model_new = RandomForestClassifier(n_estimators = 56,
                                   max_depth = 16)

model_new.fit(X_train,y_train)

model_new.score(X_test,y_test)
```
