# Building Classification Model using KNN


````python
#Import the module
from sklearn.neighbors import KNeighborsClassifier

#Split the data
X= churn_df[["total_day_charge","total_eve_charge"]].values
y=churn_df['churn'].values

#create and instantiate the model
knn= KNeighborsClassifier(n_neighbors=6)

#fit the model
knn.fit(X,y)

#create the unseen data
X_new= [[50,10],
        [150,15]]
#use the model to predict
predictions= knn.predict(X_new)

print(predictions)
````

#Measuring Model Performance using Train/test split

````python
from sklearn.model_selection import train_test_split

#split the data
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.3, random_state= 21,
                                                stratify=y)

#model
knn=KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train) #fitting training data to model

knn.score(X_test,y_test) #testing on test data to get accuracy
````

#Model Complexity and under/overfitting with value of K

````python
import matplotlib.pyplot as plt

train_accuracies= {} #to store training and test accuracies
test_accuracies= {}

neighbors= np.arange(1,25) #array of values for k

for neighbor in neighbors:
  #create the model
  knn=KNeighborsClassifier(n_neighbors=neighbor)
  #fit the model on training data
  knn.fit(X_train,y_train)

  train_accuracies[neighbor]= knn.score(X_train,y_train)
  test_accuracies[neighbor]= knn.score(X_test,y_test)

plt.figure(figsize=(6,6))
plt.title('Model Complexity')
plt.plot(neighbors,train_accuracies.values(),label='Train Accuracies')
plt.plot(neighbors, test_accuracies.values(),label='Test Accuracies')
plt.legend()
plt.xlabel('values of k')
plt.ylabel('Accuracies')
plt.show()
````

For k=16, we can see that our model has better accuracy.

In less complex model:
k(larger) causes underfitting

In more complex model:
k(small) causes overfitting.
